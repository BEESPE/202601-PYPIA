```python
from pathlib import Path

import cv2
import matplotlib.pyplot as plt
```

Charger l'image istockphoto_faces.jpg en utilisant cv2 et la convertir en niveaux de gris.

```python
color_face_array = cv2.imread(Path("../../data/harry_potter_faces.jpg"))


grey_face_array = cv2.cvtColor(
    color_face_array,
    cv2.COLOR_BGR2GRAY,
)

plt.imshow(grey_face_array, cmap="Greys_r")
plt.show()
```

Charger le classifier en instanciant un CascadeClassifier  partir du fichier xml tlcharg prcdemment.

```python
model = cv2.CascadeClassifier("../../models/haarcascade_frontalface_alt.xml")
```

L'objet ainsi cr dispose d'une mthode detectMultiScale, qui prend une image en argument et applique le CascadeClassifier dessus. MultiScale fait rfrence au fait que l'algorithme s'intresse  des sous-rgions de l'image de diffrentes chelles, afin de dtecter des visages de diffrentes tailles

Appliquer cette mthode et stocker le rsultat dans une variable.

```python
detections = model.detectMultiScale(grey_face_array)
```

La variable cre contient maintenant tous les visages dtects dans notre image. Il ne reste plus qu' les visualiser.

Pour cela, utiliser la fonction rectangle d'OpenCV. Consulter la documentation de cette fonction. Pour information, chaque visage dtect stock suite  l'appel de face_cascade est dfini par les coordonnes de son coin suprieur gauche, ainsi que la largeur et la hauteur du rectangle qui contient ce visage. Contrainte supplmentaire : on crera des rectangles verts d'paisseur 2.

```python
for x, y, delta_x, delta_y in detections:
    cv2.rectangle(
        color_face_array,
        (x, y),
        (x + delta_x, y + delta_y),
        color=(0, 255, 0),
        thickness=2,
    )
```

Pour finir, afficher l'image. Attention, il s'agit d'une image cv2 ;)

```python
cv2.imshow("image visages", color_face_array)
cv2.waitKey(0)
cv2.destroyAllWindows
```
